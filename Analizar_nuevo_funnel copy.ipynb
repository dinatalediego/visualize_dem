### Línea por línea

```python
import pandas as pd

def cargar_hojas_por_palabra(ruta_archivo: str) -> dict:
    """
    Lee todas las hojas de un archivo Excel y devuelve un dict
    donde la clave es la primera palabra del nombre de la hoja
    y el valor es el DataFrame correspondiente.
    """
    # lee todas las hojas en un dict {nombre_hoja: df}
    hojas = pd.read_excel(ruta_archivo, sheet_name=None)
    resultado = {}
    for nombre_hoja, df in hojas.items():
        # extrae la primera palabra (antes de espacios)
        clave = nombre_hoja.split()[0]
        # guarda copia del df bajo esa clave
        resultado[clave] = df.copy()
    return resultado

# Ejemplo de uso:
if __name__ == "__main__":
    ruta = path
    dfs = cargar_hojas_por_palabra(ruta)
    for clave, df in dfs.items():
        print(f"{clave}: {df.shape[0]} filas, {df.shape[1]} columnas")
```

### Oculta

```python
from typing import Dict, List, Tuple
import pandas as pd
from openpyxl import load_workbook

class ExcelLoader:
    def __init__(self, filepath: str):
        """
        Inicializa el loader con la ruta del archivo.
        """
        self.filepath = filepath
        # Cargamos el workbook una sola vez
        self._wb = load_workbook(filepath, data_only=True)
    
    def _hidden_dims(self, sheet_name: str) -> Tuple[List[int], List[str]]:
        """
        Devuelve (filas_ocultas, cols_ocultas) de la hoja dada.
        - filas_ocultas: lista de índices Excel (1-based)
        - cols_ocultas: lista de letras de columna
        """
        ws = self._wb[sheet_name]
        filas = [r for r, dim in ws.row_dimensions.items() if dim.hidden]
        cols  = [c for c, dim in ws.column_dimensions.items() if dim.hidden]
        return filas, cols
    
    def load_sheets_by_prefix(self) -> Dict[str, pd.DataFrame]:
        """
        Lee todas las hojas, elimina filas/columnas ocultas
        y las agrupa en un dict bajo la primera palabra de su nombre.
        """
        # lee todos los dfs con pandas
        all_sheets = pd.read_excel(self.filepath, sheet_name=None)
        result: Dict[str, pd.DataFrame] = {}
        
        for sheet_name, df in all_sheets.items():
            filas_ocultas, cols_ocultas = self._hidden_dims(sheet_name)
            
            # Drop hidden rows (ajustamos a 0-based)
            if filas_ocultas:
                df = df.drop(index=[r - 1 for r in filas_ocultas], errors="ignore")
            # Drop hidden columns
            if cols_ocultas:
                df = df.drop(columns=cols_ocultas, errors="ignore")
            
            key = sheet_name.split()[0]
            result[key] = df.reset_index(drop=True)
        
        return result
```

```python

# Ejemplo de uso
if __name__ == "__main__":
    loader = ExcelLoader(path)
    dfs = loader.load_sheets_by_prefix()
    for prefix, df in dfs.items():
        print(f"{prefix}: {df.shape[0]} filas × {df.shape[1]} columnas")
```

### Leyendo df con filas y col ocultas

```python
from typing import Dict, List, Tuple
import pandas as pd
from openpyxl import load_workbook

class ExcelLoader:
    def __init__(self, filepath: str):
        """
        Inicializa el loader con la ruta del archivo.
        """
        self.filepath = filepath
        self._wb = load_workbook(filepath, data_only=True)
    
    def _hidden_dims(self, sheet_name: str) -> Tuple[List[int], List[str]]:
        """
        Devuelve (filas_ocultas, cols_ocultas) de la hoja dada.
        - filas_ocultas: lista de índices Excel (1-based)
        - cols_ocultas: lista de letras de columna
        """
        ws = self._wb[sheet_name]
        filas = [r for r, dim in ws.row_dimensions.items() if dim.hidden]
        cols  = [c for c, dim in ws.column_dimensions.items() if dim.hidden]
        return filas, cols
    
    def load_sheets_by_prefix(self) -> Dict[str, pd.DataFrame]:
        """
        Lee todas las hojas, elimina filas/columnas ocultas,
        ajusta encabezados y renombra la primera columna a "MÉTRICAS",
        y las agrupa en un dict bajo la primera palabra de su nombre.
        """
        # lee todos los dfs con pandas
        all_sheets = pd.read_excel(self.filepath, sheet_name=None, header=None)
        result: Dict[str, pd.DataFrame] = {}
        
        for sheet_name, df in all_sheets.items():
            # 1) Elimina filas/columnas ocultas
            filas_ocultas, cols_ocultas = self._hidden_dims(sheet_name)
            if filas_ocultas:
                df = df.drop(index=[r - 1 for r in filas_ocultas], errors="ignore")
            if cols_ocultas:
                df = df.drop(columns=cols_ocultas, errors="ignore")
            
            # 2) Ajuste de encabezados: usar la fila 1 (segunda) como names
            df.columns = df.iloc[1]
            # 3) Eliminar las dos primeras filas (0 y 1)
            df = df.iloc[2:]#.reset_index(drop=True)
            # 4) Renombrar la primera columna a "MÉTRICAS"
            #first_col = 
            df.rename(columns={df.columns[0]: "MÉTRICAS"}, inplace=True)
            
            # 5) Guardar bajo la clave de la primera palabra
            key = sheet_name.split()[0]
            result[key] = df
        
        return result

# Ejemplo de uso
if __name__ == "__main__":
    loader = ExcelLoader(path)
    dfs = loader.load_sheets_by_prefix()
    for prefix, df in dfs.items():
        print(f"{prefix}: {df.shape[0]} filas × {df.shape[1]} columnas")
        display(df.head())  # si estás en Jupyter, para ver un vistazo
```
