### Línea por línea

```python
from typing import Dict, List, Tuple
import pandas as pd
from openpyxl import load_workbook

class ExcelLoader:
    def __init__(self, filepath: str):
        """
        Inicializa el loader con la ruta del archivo.
        """
        self.filepath = filepath
        # Cargamos el workbook una sola vez
        self._wb = load_workbook(filepath, data_only=True)
    
    def _hidden_dims(self, sheet_name: str) -> Tuple[List[int], List[str]]:
        """
        Devuelve (filas_ocultas, cols_ocultas) de la hoja dada.
        - filas_ocultas: lista de índices Excel (1-based)
        - cols_ocultas: lista de letras de columna
        """
        ws = self._wb[sheet_name]
        filas = [r for r, dim in ws.row_dimensions.items() if dim.hidden]
        cols  = [c for c, dim in ws.column_dimensions.items() if dim.hidden]
        return filas, cols
    
    def load_sheets_by_prefix(self) -> Dict[str, pd.DataFrame]:
        """
        Lee todas las hojas, elimina filas/columnas ocultas
        y las agrupa en un dict bajo la primera palabra de su nombre.
        """
        # lee todos los dfs con pandas
        all_sheets = pd.read_excel(self.filepath, sheet_name=None)
        result: Dict[str, pd.DataFrame] = {}
        
        for sheet_name, df in all_sheets.items():
            filas_ocultas, cols_ocultas = self._hidden_dims(sheet_name)
            
            # Drop hidden rows (ajustamos a 0-based)
            if filas_ocultas:
                df = df.drop(index=[r - 1 for r in filas_ocultas], errors="ignore")
            # Drop hidden columns
            if cols_ocultas:
                df = df.drop(columns=cols_ocultas, errors="ignore")
            
            key = sheet_name.split()[0]
            result[key] = df.reset_index(drop=True)
        
        return result
```

```python

# Ejemplo de uso
if __name__ == "__main__":
    loader = ExcelLoader(path)
    dfs = loader.load_sheets_by_prefix()
    for prefix, df in dfs.items():
        print(f"{prefix}: {df.shape[0]} filas × {df.shape[1]} columnas")
```
